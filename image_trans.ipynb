```python
import cv2 as cv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('//content//vv.jpg')
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(image)
```

```python
image.shape
```

```python
gray_image = 0.114*image[:,:,0] + 0.587*image[:,:,1] + 0.299*image[:,:,2]
gray_image = gray_image.astype('uint8')
```

```python
cv2_imshow(gray_image)
```

```python
gray_image.shape
```

Negative Image

```python
negative_image = 255 - gray_image
```

```python
cv2_imshow(negative_image)
```

mirror image

```python
mirror_image = gray_image[:,::-1]
cv2_imshow(mirror_image)
```

Water Image

```python
water_image = gray_image[::-1,:]
cv2_imshow(water_image)
```

Draw Outline

```python
gray_image[:,:2] = 0
gray_image[:,223:]=0
gray_image[:2,:]=0
gray_image[223:,:]=0
```

```python
cv2_imshow(gray_image)
```

Middle of image

```python
mid = gray_image
mid[:,111:113] = 0
```

```python
cv2_imshow(mid)
```
